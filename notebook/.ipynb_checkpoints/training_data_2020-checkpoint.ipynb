```python
import os
import zipfile
import pandas as pd
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
import os
import re
import shutil
```

```python
# Define the root directory and target directory
root_dir = '../data/MICCAI_BraTS2020_TrainingData'
target_extension = '.nii'
flat_dir = os.path.join('../data/flattened_nii_files_2020')

# Create the target directory if it doesn't exist
os.makedirs(flat_dir, exist_ok=True)

# Traverse through the folders starting with "MICCAI" and flatten the .nii files
for folder_name in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder_name)
    if folder_name.startswith('BraTS20_Training') and os.path.isdir(folder_path):
        for root, _, files in os.walk(folder_path):
            for file in files:
                if file.endswith(target_extension):
                    source_file = os.path.join(root, file)
                    target_file = os.path.join(flat_dir, file)
                    # Avoid overwriting files with the same name
                    counter = 1
                    while os.path.exists(target_file):
                        name, ext = os.path.splitext(file)
                        target_file = os.path.join(flat_dir, f"{name}_{counter}{ext}")
                        counter += 1
                    shutil.copy(source_file, target_file)

```

```python
# Initialize a list to store file details
file_details = []
processed_dir = '../data/flattened_nii_files_2020'

pattern = re.compile(
    r"BraTS20_Training_(\d{3})_(flair|t1|t1ce|t2|seg)(_processed_[a-zA-Z0-9]+)?\.nii"
)

file_details = []

# Loop through the directory and collect file details
for root, dirs, files in os.walk(processed_dir):
    for file_name in files:
        if file_name.endswith('.nii'):  # Check if it's an .nii file
            match = pattern.match(file_name)
            if match:
                subject_id, scan_type, processing_info = match.groups()

                # Normalize file path
                normalized_file_path = os.path.join(root, file_name).replace('\\', '/')

                # Add details to the list
                file_details.append({
                    'Subject ID': subject_id,
                    'Scan Type': scan_type,
                    'Processing Info': processing_info.lstrip('_') if processing_info else 'raw',  # Remove leading underscore if present
                    'File Path': normalized_file_path
                })
            else:
                print(f"File does not match expected structure: {file_name}")

summary_df = pd.DataFrame(file_details)

# Save or display the summary
summary_file = '../data/training_detailed_summary_2020.csv'
summary_df.to_csv(summary_file, index=False)

print(f"Detailed summary created: {summary_file}")
```

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Initialize lists to store results
blank_slice_counts = []

# Filter out segmentation files (assuming they contain 'seg' in their filenames)
filtered_file_paths = [path for path in summary_df['File Path'] if 'seg' not in path.lower()]

# Iterate over all non-segmentation .nii file paths
for nii_file in filtered_file_paths:
    try:
        # Load the .nii file
        nii_data = nib.load(nii_file)

        # Extract the image data
        image_data = nii_data.get_fdata()

        # Iterate over each slice along the z-axis
        blank_counts = []
        for slice_idx in range(image_data.shape[2]):
            # Check if the slice is entirely blank (all zeros)
            if np.all(image_data[:, :, slice_idx] == 0):
                blank_counts.append(1)
            else:
                blank_counts.append(0)
        
        # Add the blank slice counts for each index
        blank_slice_counts.append(blank_counts)

    except Exception as e:
        print(f"Error processing file {nii_file}: {e}")

# Convert to a DataFrame for easier manipulation
blank_slice_counts = np.array(blank_slice_counts)
index_counts = np.sum(blank_slice_counts, axis=0)

# Plot the histogram
plt.figure(figsize=(12, 6))
plt.bar(range(len(index_counts)), index_counts, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Slice Index")
plt.ylabel("Count of Blank Slices")
plt.title("Histogram of Blank Slices Over Slice Index Across All Non-Segmentation .nii Files")
plt.xticks(ticks=range(0, len(index_counts), 5), rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Initialize lists to store sums of pixel values for each slice
slice_pixel_sums = []

# Filter out segmentation files (assuming they contain 'seg' in their filenames)
filtered_file_paths = [path for path in summary_df['File Path'] if 'seg' not in path.lower()]

# Iterate over all non-segmentation .nii file paths
for nii_file in filtered_file_paths:
    try:
        # Load the .nii file
        nii_data = nib.load(nii_file)

        # Extract the image data
        image_data = nii_data.get_fdata()

        # Iterate over each slice along the z-axis
        for slice_idx in range(image_data.shape[2]):
            # Calculate the sum of pixel values in the slice
            pixel_sum = np.sum(image_data[:, :, slice_idx])
            slice_pixel_sums.append((nii_file, slice_idx, pixel_sum))

    except Exception as e:
        print(f"Error processing file {nii_file}: {e}")

# Convert the results to a DataFrame for easier manipulation
pixel_sums_df = pd.DataFrame(slice_pixel_sums, columns=['File Path', 'Slice Index', 'Pixel Sum'])

# Visualize the distribution of pixel sums (histogram)
plt.figure(figsize=(12, 6))
plt.hist(pixel_sums_df['Pixel Sum'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Sum of Pixel Values per Slice")
plt.ylabel("Frequency")
plt.title("Distribution of Sum of Pixel Values per Slice Across All .nii Files")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Initialize an array to store the count of non-zero pixels across all slices and subjects
# Assuming all slices have the same dimensions
slice_dimensions = None
non_zero_pixel_counts = None

# Filter out segmentation files (assuming they contain 'seg' in their filenames)
filtered_file_paths = [path for path in summary_df['File Path'] if 'seg' not in path.lower()]

# Iterate over all non-segmentation .nii file paths
for nii_file in filtered_file_paths:
    try:
        # Load the .nii file
        nii_data = nib.load(nii_file)

        # Extract the image data
        image_data = nii_data.get_fdata()

        # Initialize non_zero_pixel_counts array if not already done
        if non_zero_pixel_counts is None:
            slice_dimensions = image_data.shape[:2]
            non_zero_pixel_counts = np.zeros(slice_dimensions, dtype=int)

        # Iterate over each slice and count non-zero pixels
        for slice_idx in range(image_data.shape[2]):
            non_zero_pixel_counts += (image_data[:, :, slice_idx] != 0).astype(int)

    except Exception as e:
        print(f"Error processing file {nii_file}: {e}")

# Create a heatmap
plt.figure(figsize=(8, 8))
plt.imshow(non_zero_pixel_counts, cmap='hot', interpolation='nearest')
plt.colorbar(label='Count of Non-Zero Pixels')
plt.title("2D Heat Map of Non-Zero Pixel Counts Across All Slices and Subjects")
plt.axis("off")
plt.tight_layout()

plt.show()
```
